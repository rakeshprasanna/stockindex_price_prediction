```python
import yfinance as yf
```

```python
nifty_next_50 = yf.Ticker("^NSMIDCP")
```

```python
nifty_next_50 = nifty_next_50.history(period='max')
```

```python
nifty_next_50
```

```python
nifty_next_50.index
```

```python
nifty_next_50.plot.line(y='Close',use_index=True)
```

```python
del nifty_next_50['Dividends']
del nifty_next_50['Stock Splits']
```

```python
nifty_next_50['Tomorrow']=nifty_next_50['Close'].shift(-1)
```

```python
nifty_next_50
```

```python
nifty_next_50['Target']=(nifty_next_50['Tomorrow'] > nifty_next_50['Close']).astype(int)
nifty_next_50
```

```python
nifty_next_50=nifty_next_50.loc["1990-01-01":].copy()
nifty_next_50
```

```python
from sklearn.ensemble import RandomForestClassifier

model= RandomForestClassifier(min_samples_split=200,random_state=3)

train=nifty_next_50.iloc[:-50]
test=nifty_next_50.iloc[-50:]

predictors=['Close','Volume','High','Low']
model.fit(train[predictors],train["Target"])
```

```python
from sklearn.metrics import precision_score

preds=model.predict(test[predictors])
preds
```

```python
import pandas as pd
preds = pd.Series(preds,index=test.index)
preds
```

```python
precision_score(test['Target'],preds)
```

```python
combined=pd.concat([test["Target"],preds],axis=1)
combined.plot()
```

```python
def predict(train,test,predictors,model):
  model.fit(train[predictors],train["Target"])
  preds=model.predict(test[predictors])
  preds=pd.Series(preds,index=test.index,name="Predictions")
  combined=pd.concat([test["Target"],preds],axis=1)
  return combined
```

```python
def backtest(data,model,predictors,start=2500,step=250):
  all_predictions = []
  for i in range(start,data.shape[0],step):
    train = data.iloc[0:i].copy()
    test=data.iloc[i:(i+step)].copy()
    predictions=predict(train,test,predictors,model)
    all_predictions.append(predictions)
  return pd.concat(all_predictions)
```

```python
predictions=backtest(nifty_next_50,model,predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"],predictions["Predictions"])
```

```python
predictions["Target"].value_counts() /predictions.shape[0]
```

```python
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
  rolling_averages = nifty_next_50.rolling(horizon).mean()

  ratio_column = f"Close_Ratio_{horizon}"
  nifty_next_50[ratio_column] = nifty_next_50["Close"] / rolling_averages["Close"]

  trend_column = f"Trend_{horizon}"
  nifty_next_50[trend_column]=nifty_next_50.shift(1).rolling(horizon).sum()["Target"]

  new_predictors +=[ratio_column,trend_column]
```

```python
nifty_next_50=nifty_next_50.dropna()
nifty_next_50
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50,random_state=1)
```

```python
def predict(train,test,predictors,model):
  model.fit(train[predictors],train["Target"])
  preds=model.predict_proba(test[predictors])[:,1]
  preds[preds>=.60]=1
  preds[preds<=.6]=0
  preds=pd.Series(preds,index=test.index,name="Predictions")
  combined=pd.concat([test["Target"],preds],axis=1)
  return combined
```

```python
predictions = backtest(nifty_next_50,model,new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"],predictions["Predictions"])
```
