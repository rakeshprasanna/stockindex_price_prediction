```python
import yfinance as yf
```

```python
nifty_next_50 = yf.Ticker("^NSMIDCP")
```

```python
nifty_next_50 = nifty_next_50.history(period='max')
```

```python
nifty_next_50.plot.line(y='Close',use_index=True)
```

```python
del nifty_next_50['Dividends']
del nifty_next_50['Stock Splits']
```

```python
nifty_next_50['Tomorrow']=nifty_next_50['Close'].shift(-1)
```

```python
nifty_next_50
```

```python
nifty_next_50['Target']=(nifty_next_50['Tomorrow']>nifty_next_50['Çlose']).astype(int)
nifty_next_50
```
